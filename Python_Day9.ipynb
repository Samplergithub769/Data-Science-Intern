**Pandas Date Range**

```python
# Year-Month-Date->Days Frequency
pd.date_range(start='2025-02-02',end='2025-02-10')
```

```python
# Months Frequency
pd.date_range(start='2025-02-02',end='2025-09-10',freq='M')
```

```python
# Weekdays Frequency
pd.date_range(start='2025-02-03',end='2025-02-10',freq='B') #B indicates business days Monday to Friday
```
