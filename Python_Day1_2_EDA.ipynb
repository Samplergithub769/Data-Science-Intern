**Handling Outliers using IQR**

```python
import numpy as np
```

```python
list_=[1,2,3,4,5,6,9,10,12,55,60]
```

```python
Q1=np.percentile(list_,25)
Q1
```

```python
Q3=np.percentile(list_,75)
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
low=Q1-1.5*IQR
upp=Q3+1.5*IQR
```

```python
low,upp
```

```python
outliers=[]
```

```python
list_
```

```python
for i in list_:
  if((i>upp)or(i<low)):
    outliers.append(i)
```

```python
outliers
```

```python
for i in list_:
  if i not in outliers:
    print(i)
```

```python
list_=[i for i in list_ if i not in outliers]
```

```python
list_
```
