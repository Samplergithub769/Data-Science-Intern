Pandas Series

*   A pandas series is like a column in a table
*   It is a one-dimensional array holding data of any type




```python
import pandas as pd
```

```python
#Series Method

series=pd.Series([1,2,3,4,5,6,7])
print(series)
```

```python
#Series method with index
series=pd.Series([1,2,3,4,5],index=["a","b","c","d","e"])
print(series)
```

```python
#Acessing Series elements
print(series["a"])
print(series["b"])
```

**Pandas Dataframe**
A Data frame is a two-dimensional data structure,i.e,data is aligned in a tabular fashion in rows and columns.

```python
#DataFrame method in list
data=pd.DataFrame([1,2,3,4,5])
print(data)
```

```python
#DataFrame using dictionary
data=pd.DataFrame({"Name":['Shaily','Deepali','Ansh','Harsita','Manshi','Sneha','Yashi'],"Age":[12,13,14,15,16,17,18]})
print(data)
data.head()
```

```python
#DataFrame with index
data=pd.DataFrame({"Name":['Shaily','Deepali','Ansh','Harsita','Manshi','Sneha','Yashi'],"Age":[12,13,14,15,16,17,18]},index=["C1","C2","C3","C4","C5","C6","C7"])
print(data)
data.head()
```

```python
#Column Selection
data["Name"]
```

```python
data["Age"]
```

```python
data.head(10)
```

```python
#Row Selection
data.loc["C1","Name"]
```

```python
#Row selection
data.iloc[2,0]
```

**Pandas Concatenation**

It is used to concatenate pandas objects along a particular axis with optional set logic along the other axes.


```python
names=pd.DataFrame({"Name":["Shaily","Ansh","Deepali","Manshi","Harsita","Sneha","Yashi"]})
age=pd.DataFrame({"Age":[12,13,14,15,16,17,18]})
```

```python
names
```

```python
age
```

```python
pd.concat([names,age],axis=1)
```

```python
names2=pd.DataFrame({"Name":["Moti","Shalu","Shubh","Gudiya","Anvee","Suman","Pooja"]})
names2
```

```python
pd.concat([names,names2],axis=0)
```
